```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
data['discounted_price'] = data['discounted_price'].astype(str)
data['rating'] = data['rating'].astype(str)
data['rating_count'] = data['rating_count'].astype(str)
```

```python
data['discounted_price'] = pd.to_numeric(data['discounted_price'].str.replace(',', '').str.strip(), errors='coerce')
data['rating'] = pd.to_numeric(data['rating'], errors='coerce')
data['rating_count'] = pd.to_numeric(data['rating_count'], errors='coerce')
```

```python
print(data[['discounted_price', 'rating', 'rating_count']].isnull().sum())
```

```python
##According to the table above, the data is not cleaned properly considering rating and rating count
```

```python
data_cleaned = data.dropna(subset=['rating', 'rating_count'])
```

```python
data['rating'] = data['rating'].fillna(data['rating'].mean())
data['rating_count'] = data['rating_count'].fillna(data['rating_count'].mean())
```

```python
print(data[['discounted_price', 'rating', 'rating_count']].isnull().sum())
```

```python
features = data_cleaned[['discounted_price', 'rating', 'rating_count']]
```

```python
features = features.dropna()
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
scaled_features_df = pd.DataFrame(scaled_features,columns=features.columns)
```

```python
inertia=[]
```

```python
k_values=range(1,11)
```

```python
for k in k_values:
    kmeans=KMeans(n_clusters=k,random_state=42)
    kmeans.fit(scaled_features_df)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(8,6))
plt.plot(k_values, inertia, marker='o')
plt.title('Elbow Method for optimal k')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.grid()
plt.show()
```

```python
## According to the Graph above using elbow method we can see that upto 3 clusters it is showing no changes(straight line),
##so it is optimal to take k as 3
```

```python
optimal_k = 3
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
data_cleaned.loc[:, 'Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
print(data_cleaned[['product_id','discounted_price','rating','rating_count','Cluster']].head())
```

```python
data_cleaned.columns ##making sure of the data columns 
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data_cleaned['discounted_price'],data_cleaned['rating'],c=data_cleaned['Cluster'],cmap='viridis',alpha = 0.6)
plt.title('Clustering of Products on discounted price and rating')
plt.xlabel('Discounted_price')
plt.ylabel('Rating')
plt.colorbar(label = 'Cluster')
plt.show()
```

```python
## The colors assigned to the points correspond to the different clusters you identified using K-Means clustering. 
##Each color represents a unique cluster.
##The color gradient (using the viridis colormap) provides a visual reference to differentiate clusters.
```

```python
## From the table we can see,
##Cluster 1 is situated in the upper right quadrant,
##it indicates that these products generally have high ratings and high discounted prices. If higher prices tend to align with higher ratings, 
##it suggests that consumers are willing to pay more for products they believe are of higher quality.
```

```python
cluster_summary = data_cleaned.groupby('Cluster').agg({
    'discounted_price': ['mean','median','max','min'],
    'rating' : ['mean','count']
}).reset_index()
cluster_summary.columns = ['Cluster', 'Mean_Discounted_Price', 'Median_Discounted_Price', 
                           'Max_Discounted_Price', 'Min_Discounted_Price', 
                           'Mean_Rating', 'Count']
print(cluster_summary)
```

```python
##According to the table,
##Cluster 0: Focuses marketing efforts on affordability and value. It Highlights discounts and user reviews to attract budget-conscious consumers.
##Cluster 1: Uses premium branding and high quality and exclusivity. Targets high-income consumers and consider collaborations with influencers 
##in luxury markets.
##Cluster 2: Promotes these products as high-quality options at reasonable prices. Emphasizes value and quality in marketing campaigns.
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Cluster',y='discounted_price',data= data_cleaned)
plt.title('Box plot of discounted price on clusters')
plt.xlabel('Cluster')
plt.ylabel('Discounted price')
plt.show()
```
